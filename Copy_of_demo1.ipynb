## Part 4: Missing value

### Step 1: Find column with null values and display as list

```python
columns_with_null = merged_df.columns[merged_df.isnull().any()].tolist()

print("Columns with null values:")
print(columns_with_null)
```

### Step 2: Locate with null values in column

```python
missing_age = merged_df[merged_df['Age'].isnull()]
missing_age
```

### Step 3: Fill in the missing value

```python
merged_df.at[3, 'Age'] = 57
merged_df.at[5, 'Age'] = 23
merged_df.at[8, 'Age'] = 26
merged_df.head(10)
```

```python
merged_df.at[8, 'Department'] = 'MARKETING'
merged_df.at[2, 'Job Position'] = 'FINANCIAL ANALYST'
merged_df.head(10)
```

## Part 5: Inconsistent data formatting - date



### Step 1: Convert a column of Dataframe to datatime object

```python
# ISO 8601 date fromat: â€œYYYY-MM-DDâ€
merged_df['Date of Joining'] = pd.to_datetime(merged_df['Date of Joining'], errors='coerce')
merged_df.head(10)
```

### Step 2: Check data type

```python
merged_df.dtypes
```

### Other date style

```python
merged_df['Date of Joining'] = merged_df['Date of Joining'].dt.strftime("%Y/%m/%d")
merged_df
```

## Part 6: Inconsistent data formatting - change case

### Step 1: Change to all upper case



*   Convert to uppercase: upper()
*   Convert to lowercase: lower()
*   Capitalize string: capitalize()






```python
merged_df['First Name'] = merged_df['First Name'].str.upper()
merged_df['Last Name'] = merged_df['Last Name'].str.upper()
merged_df['Gender'] = merged_df['Gender'].str.upper()
merged_df['Department'] = merged_df['Department'].str.upper()
merged_df['Job Position'] = merged_df['Job Position'].str.upper()
merged_df.head(10)
```

## Part 7: Inconsistent column names / values

### Step 1: Count the value of column 'Gender'

```python
merged_df['Gender'].value_counts()
```

### Step 2: Convert the values starting with 'F' to 'FEMALE' and values starting with 'M' to 'MALE'

```python
merged_df['Gender'] = merged_df['Gender'].apply(lambda x: 'MALE' if x.startswith('M') else 'FEMALE' if x.startswith('F') else x)
merged_df.head(10)
```

```python
merged_df['Gender'].value_counts()
```

## Part 8: Duplicate

### Step 1: Check duplicated row

```python
merged_df[merged_df.duplicated(keep=False)]
```

### Step 2: Drop duplicated row

```python
merged_df = merged_df.drop_duplicates()
merged_df.head(10)
```

```python
merged_df = merged_df.reset_index(drop=True)
merged_df.head(10)
```

## Part 9: Lending /Trailing whitespace

### Step 1: Check whitespace

```python
for column in merged_df.columns:
    if (merged_df[column].astype(str).str.startswith(' ') | merged_df[column].astype(str).str.endswith(' ')).any():
        print(f"Column '{column}' has leading or trailing whitespace.")
```

```python
whitespace = merged_df['Email'].str.startswith(' ') | merged_df['Email'].str.endswith(' ')
whitespace
```

```python
merged_df['Email'][3]
```

### Step 2: Remove leading and trailing whitespace

*   remove both leading and trailing whitespace: strip()
*   remove only leading whitespace: lstrip()
*   remove only trailing whiteshpace: rstrip()



```python
merged_df['Email'] = merged_df['Email'].str.strip()
merged_df
```

```python
merged_df['Email'][3]
```
